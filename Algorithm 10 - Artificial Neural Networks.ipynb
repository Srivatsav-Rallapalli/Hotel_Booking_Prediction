```python
import pandas as pd

import warnings
warnings.simplefilter('ignore')
```

```python
# Importing the Dataset

df = pd.read_csv(r'clean.csv')
df
```

```python
# Seperating the Dependent and Independent variables

X = df.drop(columns=['booking_status'])
y = df['booking_status']
```

```python
# Balancing the Dataset
from imblearn.under_sampling import RandomUnderSampler

rs = RandomUnderSampler(random_state=17)
x_new,y_new = rs.fit_resample(X,y)
```

```python
# Splitting the Dataset into Train and Test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_new,y_new,random_state=17,train_size=0.8)
```

```python
# Importing the necessary libraries for ANN
# Initializing the model

import tensorflow as tf
import keras
from keras.models import Sequential

cnn = Sequential()
```

```python
# Adding the Input Layer and 1st Hidden Layer to the model
from keras.layers import Dense

cnn.add(Dense(input_dim=24,units=48,kernel_initializer='uniform',activation='relu'))

# Adding the Second Hidden Layer
cnn.add(Dense(units=48,kernel_initializer='uniform',activation='relu'))

# Adding the Third Hidden Layer
cnn.add(Dense(units=48,kernel_initializer='uniform',activation='relu'))
```

```python
# Adding the output layer

cnn.add(Dense(units=1,kernel_initializer='uniform',activation='relu'))
```

```python
# Compiling the model

cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Train the model on train dataset

cnn.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
# Predicting and Evaluating the model

ypred = cnn.predict(x_test)
ypred = ypred>0.5
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print('Test Accuracy is :',accuracy_score(y_test,ypred))
```

```python
confusion_matrix(y_test,ypred)
```

```python
# Cross validating the model 

def cnn_validation():
    classifier = Sequential()
    classifier.add(Dense(input_dim=24,units=48,kernel_initializer='uniform',activation='relu'))
    classifier.add(Dense(units=48,kernel_initializer='uniform',activation='relu'))
    classifier.add(Dense(units=48,kernel_initializer='uniform',activation='relu'))
    classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
from scikeras.wrappers import KerasClassifier

classifier = KerasClassifier(cnn_validation(),batch_size=32,epochs=100)
```

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(classifier,x_train,y_train,cv=5)
```
