```python
import pandas as pd

import warnings
warnings.simplefilter('ignore')
```

```python
# Importing the Dataset

df = pd.read_csv(r'clean.csv')
df
```

```python
# Seperating the Dependent and Independent variables

X = df.drop(columns=['booking_status'])
y = df['booking_status']
```

```python
# Balancing the Data
from imblearn.under_sampling import RandomUnderSampler

rs = RandomUnderSampler(random_state=17)
x_new,y_new = rs.fit_resample(X,y)
```

```python
# Splitting the data into Train and Test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_new,y_new,random_state=17,train_size=0.8)
```

```python
# Importing necessary libraries for ANN
# Initialize the Sequential model

import tensorflow as tf
import keras
from keras.models import Sequential

ann = Sequential()
```

```python
from keras.layers import Dense

ann.add(Dense(input_dim=24, units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
ypred = ann.predict(x_test)
ypred = ypred>0.5
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

confusion_matrix(y_test,ypred)
```

```python
print('Test Accuracy is :',accuracy_score(y_test,ypred))
```

```python
def cross_validator():
    classifier = Sequential()
    classifier.add(Dense(input_dim=24,units=1,kernel_initializer='uniform',activation='sigmoid'))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
from scikeras.wrappers import KerasClassifier

classifier = KerasClassifier(cross_validator(),epochs=100,batch_size=32)
```

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(classifier,x_train,y_train,cv=5)
```

```python
print(accuracies)
accuracies.mean()
```
